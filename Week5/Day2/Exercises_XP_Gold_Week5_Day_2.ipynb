**Exercise 2 : Statistical Distribution Analysis**

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import skew, kurtosis

# Step 1: Generate dataset following a normal distribution
# mean = 0, standard deviation = 1, sample size = 1000
data = np.random.normal(loc=0, scale=1, size=1000)

# Step 2: Plot histogram
plt.hist(data, bins=30, edgecolor='black', density=True)
plt.title('Histogram of Normally Distributed Data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()

# Step 3: Calculate skewness and kurtosis
skewness = skew(data)
kurt = kurtosis(data)

print("Skewness:", skewness)
print("Kurtosis:", kurt)

# Step 4: Interpretation
if abs(skewness) < 0.5:
    skew_interp = "The distribution is approximately symmetric."
elif skewness > 0:
    skew_interp = "The distribution is positively skewed (longer tail on the right)."
else:
    skew_interp = "The distribution is negatively skewed (longer tail on the left)."

if abs(kurt) < 1:
    kurt_interp = "The distribution has kurtosis close to normal (mesokurtic)."
elif kurt > 0:
    kurt_interp = "The distribution is leptokurtic (sharper peak, heavier tails)."
else:
    kurt_interp = "The distribution is platykurtic (flatter peak, lighter tails)."

print("Interpretation:")
print(skew_interp)
print(kurt_interp)
```

**Exercise 3 : Time Series Forecasting**

```python
import numpy as np
from sklearn.linear_model import LinearRegression

# Generating monthly sales data as a NumPy array
monthly_sales_data = np.random.randint(100, 500, size=12)  # Random sales data for 12 months
months = np.arange(1, 13).reshape(-1, 1)  # Months as features (1 to 12)
# Step 2: Fit Linear Regression model
model = LinearRegression()

model.fit(months, monthly_sales_data)
# Step 3: Forecast next 6 months (months 13 to 18)
future_months = np.arange(13, 19).reshape(-1, 1)
future_sales = model.predict(future_months)

# Step 4: Visualization
plt.figure(figsize=(8,5))
plt.scatter(months, monthly_sales_data, color='blue', label='Actual Sales')
plt.plot(months, model.predict(months), color='red', label='Regression Line')
plt.scatter(future_months, future_sales, color='green', label='Forecasted Sales')
plt.plot(future_months, future_sales, color='green', linestyle='--')
plt.xlabel("Month")
plt.ylabel("Sales")
plt.title("Monthly Sales Forecast using Linear Regression")
plt.legend()
plt.show()

# Step 5: Print forecasted values
print("Forecasted sales for months 13-18:", future_sales)
```

**Exercise 5 : Complex Data Visualization**

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate sample data using NumPy
x = np.linspace(0, 10, 50)
y1 = np.sin(x)            # Line plot data
y2 = np.cos(x)            # Scatter plot data
categories = ['A', 'B', 'C', 'D']
values = np.random.randint(10, 100, size=4)  # Bar plot data

# Create a multi-plot layout (2 rows, 2 columns)
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Line plot
axs[0, 0].plot(x, y1, color='blue')
axs[0, 0].set_title('Line Plot: y = sin(x)')

# Scatter plot
axs[0, 1].scatter(x, y2, color='red')
axs[0, 1].set_title('Scatter Plot: y = cos(x)')

# Bar plot
axs[1, 0].bar(categories, values, color='green')
axs[1, 0].set_title('Bar Plot: Random Values')

# Combined plot (line + scatter)
axs[1, 1].plot(x, y1, label='sin(x)', color='blue')
axs[1, 1].scatter(x, y2, label='cos(x)', color='red')
axs[1, 1].set_title('Combined Line & Scatter')
axs[1, 1].legend()

# Adjust layout
plt.tight_layout()
plt.show()
```
