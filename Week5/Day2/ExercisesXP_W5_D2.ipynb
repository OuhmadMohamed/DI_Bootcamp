Exercise 1: Matrix Operations

```python
import numpy as np

# Create a 3x3 matrix
matrix = np.array([
    [2, 1, 3],
    [1, 0, 2],
    [4, 1, 8]
])

print("Matrix:")
print(matrix)

# Determinant
det = np.linalg.det(matrix)
print("\nDeterminant:")
print(det)

# Inverse
if det != 0:
    inverse = np.linalg.inv(matrix)
    print("\nInverse:")
    print(inverse)
else:
    print("Matrix is not invertible.")
```

Explanation
* np.linalg.det() calculates the determinant.
* np.linalg.inv() calculates the inverse if the determinant is not zero.

Exercise 2: Statistical Analysis

```python
import numpy as np

# Generate 50 random numbers
data = np.random.rand(50)

# Mean
mean = np.mean(data)

# Median
median = np.median(data)

# Standard deviation
std_dev = np.std(data)

print("data:", data)
print("Mean:", mean)
print("Median:", median)
print("Standard Deviation:", std_dev)
```

Explanation
* Mean = average value.
* Median = middle value.
* Standard deviation = spread of data.

Exercise 3: Date Manipulation

```python
import numpy as np

# January 2023 dates
dates = np.arange(
    np.datetime64('2023-01-01'),
    np.datetime64('2023-02-01')
)

print("Original Dates:")
print(dates)

# Convert format YYYY/MM/DD
formatted_dates = np.datetime_as_string(
    dates,
    unit='D'
)

formatted_dates = np.char.replace(
    formatted_dates,
    '-',
    '/'
)

print("\nFormatted Dates:")
print(formatted_dates)
```

Exercise 6: Basic Hypothesis Testing

```python
import numpy as np

# Productivity before training
productivity_before = np.random.normal(
    loc=50,
    scale=10,
    size=30
)

# Productivity after training
productivity_after = productivity_before + np.random.normal(
    loc=5,
    scale=3,
    size=30
)
```

Hypotheses

* Null Hypothesis (H₀)
Training does not improve productivity.
Mean After ≤ Mean Before

* Alternative Hypothesis (H₁)
Training improves productivity.
Mean After > Mean Before

```python
#Test
mean_before = np.mean(productivity_before)
mean_after = np.mean(productivity_after)

print("Mean Before:", mean_before)
print("Mean After:", mean_after)

difference = mean_after - mean_before

print("Average Improvement:", difference)

if difference > 0:
    print("Evidence suggests productivity improved.")
else:
    print("No evidence of improvement.")
```

Exercise 8: Time Series Data Manipulation

```python
import numpy as np

dates = np.arange(
    np.datetime64('2023-01-01'),
    np.datetime64('2024-01-01')
)
```

```python
#January–March
q1 = dates[
    (dates >= '2023-01-01') &
    (dates < '2023-04-01')
]

print(q1)
```

```python

#April–June
q2 = dates[
    (dates >= '2023-04-01') &
    (dates < '2023-07-01')
]

print(q2)
#July–September
q3 = dates[
    (dates >= '2023-07-01') &
    (dates < '2023-10-01')
]

print(q3)
#October–December
q4 = dates[
    (dates >= '2023-10-01') &
    (dates < '2024-01-01')
]

print(q4)
```

Exercise 10: Basic Visualization


```python
import numpy as np
import matplotlib.pyplot as plt

# Generate data
data = np.random.randint(
    1,
    100,
    20
)

# Plot
plt.figure(figsize=(8,5))

plt.plot(
    data,
    marker='o'
)

plt.title("Random Numbers Line Graph")
plt.xlabel("Index")
plt.ylabel("Value")

plt.grid(True)

plt.show()
```

Explanation
* plt.plot() creates a line chart.
* marker='o' displays points.
* grid(True) improves readability.
