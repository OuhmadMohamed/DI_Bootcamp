**Task1. Understanding LLM Evaluation**

1.1 Evaluating LLMs is more complex than traditional software because:
- Traditional software has deterministic outputs (same input → same output).
- LLMs generate probabilistic and variable outputs, which can differ each time.
- LLMs deal with natural language, which is ambiguous, creative, and context-dependent.
- There is often no single correct answer, especially for open-ended tasks.
- LLM behavior can change depending on prompt phrasing and hidden biases.

1.2 LLM safety must be evaluated to:
- Prevent generation of harmful, biased, or misleading content.
- Reduce risks of hallucinations and misinformation.
- Ensure compliance with ethical and legal standards.
- Protect users from manipulation, hate speech, or unsafe advice.
- Detect vulnerabilities that could be exploited through malicious prompts.

1.3 Adversarial testing:
- Uses tricky or malicious prompts to expose weaknesses.
- Helps identify failure modes such as:(factual errors, unsafe responses, bias, prompt injection)
- Improves robustness by retraining or refining models using these failure cases.
- Simulates real-world misuse scenarios.

1.4.1 Automated metrics (BLEU, ROUGE, Perplexity):
- Measure surface-level similarity or probability.
- Do not capture:(meaning,
coherence, creativity, tone, factual correctness)
- Can reward shallow word overlap instead of true understanding.

1.4.2 Human evaluation:
- Can judge:(fluency, relevance, logic, appropriateness)
- But is:(expensive, slow, subjective, inconsistent)

Both are complementary rather than interchangeable.

**Task2. Applying BLEU and ROUGE Metricsxxxxxxxxx**

2.1 BLEU Score (approximate reasoning)

Reference:

>Despite the increasing reliance on artificial intelligence in various industries, human oversight remains essential to ensure ethical and effective implementation.

Generated:

>Although AI is being used more in industries, human supervision is still necessary for ethical and effective application.

BLEU measures n-gram overlap.

>Common overlapping words/phrases:(industries, human, ethical, effective, (semantic matches but not lexical: reliance vs used, oversight vs supervision))

Because BLEU only counts exact matches, the overlap is moderate.

> Approximate BLEU score: ~0.35-0.45

The score is not high despite the meaning being very similar.


**2.2 ROUGE Score (approximate reasoning)xxx**
Reference:

>In the face of rapid climate change, global initiatives must focus on reducing carbon emissions and developing sustainable energy sources to mitigate environmental impact.

Generated:

>To counteract climate change, worldwide efforts should aim to lower carbon emissions and enhance renewable energy development.

ROUGE measures recall-based overlap.

>Overlapping concepts/words:(climate change, carbon emissions, global / worldwide, energy development, reduce / lower)

> Approximate ROUGE score: ~0.4-0.5

Again, meaning is preserved but wording differs.

2.3 Limitations of BLEU and ROUGE for creative/context-sensitive text

Limitations:

>They rely on exact word overlap.

>Penalize:(paraphrasing, synonyms, creative wording)

>Do not measure:(factual correctness, coherence, tone, intent)

>Not suitable for:(storytelling, dialogue, reasoning tasks, open-ended answers)

2.4 Improvements / Alternative evaluation methods

Better alternatives:

>**BERTScore:** measures semantic similarity using embeddings.

>**Human evaluation:** fluency, relevance, correctness.

>**LLM-as-a-judge:** model-based evaluation.

>**Task-specific metrics** (QA accuracy, factuality checks).

>**Adversarial testing** for robustness.

>**Hybrid approach:** automated + human review.


**Task3. Perplexity Analysis**

3.1 Compare perplexity of two models

Model A assigns probability = 0.8
Model B assigns probability = 0.4

Perplexity is inversely related to probability:

Perplexity=1/(P(word))
Model A: 1/0.8=1.25
Model B: 1/0.4=2.5

> *Model A has lower perplexity*, meaning it is more confident and better at predicting the word “mitigation.”

3.2 Perplexity = 100 (implications)

A perplexity score of 100 means:
- The model is highly uncertain.
- Predictions are poor and unreliable.
- Language generation is likely incoherent or incorrect.

Ways to improve:

- Train on larger and higher-quality datasets.
- Fine-tune on domain-specific data.
- Increase model size.
- Improve tokenization.
- Use better architectures.
- Reduce noise in training data.

**Task6. Comparative Analysis of Evaluation Methods**

Chosen NLP Task: Machine Translation
Metrics compared:
>1. BLEU
- Measures n-gram overlap.
- Fast and cheap.
- Weak for paraphrases.
- Ignores semantics.

>2. ROUGE
- Recall-oriented overlap.
- Useful for summaries.
- Still lexical, not semantic.

>3. BERTScore
- Uses embeddings.
- Captures meaning similarity.
- Better for paraphrasing.
- Computationally heavier.

>4. Human Evaluation
- Judges fluency, adequacy, and correctness.
- Most reliable.
- Expensive and subjective.

Most appropriate metric for machine translation

Best choice: Combination of BERTScore + Human Evaluation

Reason:
- BERTScore captures semantic similarity.
- Human evaluation captures fluency and correctness.
- BLEU alone is insufficient for modern translation systems.
