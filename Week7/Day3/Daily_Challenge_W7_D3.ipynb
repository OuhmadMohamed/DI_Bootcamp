# **Daily Challenge: Classifying Handwritten Digits with CNNs**

# 1. Load MNIST Dataset

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.datasets import mnist
from tensorflow.keras.utils import to_categorical

# Load data
(x_train, y_train), (x_test, y_test) = mnist.load_data()

print("Train:", x_train.shape, y_train.shape)
print("Test:", x_test.shape, y_test.shape)
```

# 2. Preprocess for Fully Connected Network

```python
# Flatten images
x_train_fc = x_train.reshape(60000, 784).astype("float32") / 255.0
x_test_fc  = x_test.reshape(10000, 784).astype("float32") / 255.0

# One-hot encode labels
y_train_oh = to_categorical(y_train)
y_test_oh  = to_categorical(y_test)
```

# 3. Fully Connected Neural Network

```python
from tensorflow.keras import models, layers

fc_model = models.Sequential([
    layers.Dense(128, activation="relu", input_shape=(784,)),
    layers.Dense(64, activation="relu"),
    layers.Dense(10, activation="softmax")
])

fc_model.compile(
    optimizer="adam",
    loss="categorical_crossentropy",
    metrics=["accuracy"]
)

fc_model.fit(x_train_fc, y_train_oh, epochs=5, batch_size=32)

fc_loss, fc_acc = fc_model.evaluate(x_test_fc, y_test_oh)
print("FC Accuracy:", fc_acc)
```

# 4. Preprocess for CNN

```python
# Reshape for CNN
x_train_cnn = x_train.reshape(-1, 28, 28, 1).astype("float32") / 255.0
x_test_cnn  = x_test.reshape(-1, 28, 28, 1).astype("float32") / 255.0
```

# 5. Convolutional Neural Network

```python
cnn_model = models.Sequential([
    layers.Conv2D(32, (3,3), activation="relu", input_shape=(28,28,1)),
    layers.MaxPooling2D((2,2)),

    layers.Conv2D(64, (3,3), activation="relu"),
    layers.MaxPooling2D((2,2)),

    layers.Flatten(),
    layers.Dense(64, activation="relu"),
    layers.Dense(10, activation="softmax")
])

cnn_model.compile(
    optimizer="adam",
    loss="categorical_crossentropy",
    metrics=["accuracy"]
)

cnn_model.fit(x_train_cnn, y_train_oh, epochs=5, batch_size=32)

cnn_loss, cnn_acc = cnn_model.evaluate(x_test_cnn, y_test_oh)
print("CNN Accuracy:", cnn_acc)
```

# 6. Compare Performance

The convolutional neural network achieves higher accuracy than the fully connected neural network. This is because CNNs preserve spatial information in images using convolutional filters, allowing them to detect patterns such as edges and shapes. In contrast, the fully connected model flattens the image, losing spatial structure. As a result, CNNs are better suited for image classification tasks and generalize more effectively.
